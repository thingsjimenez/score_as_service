### Librerias

```python
import pandas as pd
from sklearn import preprocessing
import pickle
```

### Cargar datos en bruto

```python
aprobados = pd.read_csv('data/aprobados.csv')
aprobados['pasa'] = 1
negados = pd.read_csv('data/negados.csv')
negados['pasa'] = 0
df = pd.concat([aprobados, negados], ignore_index=True)
```

### Eliminar columnas que no son necesarias

```python
eliminar = ['Unnamed: 0', 'Unnamed: 0.1', 'Total activos', 'ConoCliente', 'Fecha_corte', 'Ocupacion', 'Agencia', 'Cargo', 'Ciiu', 'Estado', 'EndeuIndi', 'Califica', 'FecIngEmp', 'FecUltAct', 'CuidadRes']
```

```python
def eliminar_columnas(df, eliminar):
    dff = df.copy()
    for ii in eliminar:
        dff.pop(ii)
    return dff
```

```python
dff = eliminar_columnas(df, eliminar)
```

```python
dff
```

### Configurar columnas con variables tipo fecha

```python
date_today = '20200815'
dff['FecIngreso'] = pd.to_datetime(df['FecIngreso'])
dff['NumDias'] = (pd.to_datetime(date_today, format='%Y%m%d', errors='ignore') - pd.to_datetime(dff['FecIngreso'])).dt.days
```

```python
dff.head(3)
```

### Organizar formato de la variable cuota

```python
bad_chars = ['$', ' ', '.']
dff['Cuota'] = dff['Cuota'].apply(lambda price: float(''.join(ii for ii in price if not ii in bad_chars)))
```

### Obtener nombres de las columnas categoricas

```python
def variables_categoricas(lista_base, lista_remover):
    categoricas = lista_base
    for ii in lista_remover:
        categoricas.remove(ii)
    return categoricas
```

```python
var_categoricas = variables_categoricas(list(dff.columns), list(dff.describe().columns))
var_categoricas.remove('FecIngreso')
var_categoricas
```

```python
dff[var_categoricas]
```

# Procesar variables categoricas

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html

```python
categ = []
def_df = dff.copy()
for ii in var_categoricas:
    le = preprocessing.LabelEncoder()
    le.fit(list(def_df[ii].unique()))
    def_df[ii] = le.transform(def_df[ii])
    categ.append(dict(zip(le.classes_, le.transform(list(le.classes_)))))
    
```

```python
def_df.pop('FecIngreso')
def_df
```

### Diccionario de las categorias

```python
dictjs = dict(zip(var_categoricas, categ))
```

```python
with open("data/dictjs.pkl", "wb") as handle:
    pickle.dump(dictjs, handle)
```

### Normalizar variables

```python
normalized_df=(def_df-def_df.min())/(def_df.max()-def_df.min())
normalized_df
```

```python
minmax = def_df.agg(['min', 'max'])
minmax
```
